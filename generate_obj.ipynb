```python
path = './objects/'
def toObj(vertices, faces, filename='hand.obj'):
    with open(path + filename, 'w') as f:
        for v in vertices:
            f.write('v %f %f %f\n' % (v[0], v[1], v[2]))

        for face in faces:
            f.write('f %d %d %d\n' % (face[0] + 1, face[1] + 1, face[2] + 1))
```

```python
import torch
from manopth.manolayer import ManoLayer

ncomps = 45
mano_layer = ManoLayer(mano_root='./mano/models', use_pca=False, ncomps=ncomps, flat_hand_mean=True)
shapes = torch.zeros(1, 10)
```

```python
import json

result_dict = {}
with open('./results/grasp_glass/param.json', 'r') as f:
    result_dict = json.load(f)

print(result_dict)
```

```python
# poses = result_dict['epoch_19']['poses']
# global_translation = result_dict['epoch_19']['global_translation']
# poses = torch.tensor(poses)
# global_translation = torch.tensor(global_translation)
poses = torch.zeros(1, ncomps + 3, dtype=torch.float32)
global_translation = torch.zeros(1, 3, dtype=torch.float32)
poses[0][0] = torch.pi / 3
poses[0][1] = torch.pi / 3


vertices, _, _ = mano_layer(poses, shapes)

vertices = vertices[0] * 12
vertices = vertices + global_translation

toObj(vertices, mano_layer.th_faces, filename='hand_flat_x_1_y_1.obj')
```
